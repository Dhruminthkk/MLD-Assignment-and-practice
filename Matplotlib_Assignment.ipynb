**Problem 1 ) We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below. <br><br>
Task:<br>
1.fitting it to the periodic function<br>
2.plot the fit<br><br>
Data<br>
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25 <br>
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
months = [i for i in range(12)]
```

```python
z = np.polyfit( months, Max, 4)
f = np.poly1d(z)
```

```python
w = np.polyfit( months, Min, 4)
f1 = np.poly1d(w)
```

```python
month_new = np.linspace(months[0], months[-1], 40)
Max_new = f(month_new)
Min_new = f1(month_new)
```

```python
plt.plot(months,Max,'o', month_new, Max_new)
plt.plot(months,Min,'o', month_new, Min_new)
```

```python
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
```

```python
titanic.dropna(how='all',inplace=True)
```

```python
counts = titanic["sex"].value_counts()
categories = ["Male","Female"]
```

```python
plt.pie(counts, labels=categories,  autopct='%.2f%%')
# plt.title('')
plt.show()
```

```python
df = titanic[["survived","sex"]].groupby("sex").sum()
df
```

```python
plot = df.plot.pie(y="survived",figsize=(8,5),startangle=90,autopct='%.2f%%')
```
