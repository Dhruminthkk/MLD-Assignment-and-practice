# Data Visualization Assignment

**In this assignment we have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
pca = decomposition.PCA(n_components=3)
X_reduced = pca.fit_transform(iris.data)
Y = iris.target
```

```python
figure = plt.figure(1,figsize=(10,6))
graph = Axes3D(figure,elev=-150,azim=110)
graph.scatter(X_reduced[:,0],X_reduced[:,1],X_reduced[:,2],c=Y,cmap=plt.cm.brg_r)
graph.set_title("Transformed dimensions with PCA")
graph.set_xlabel("First eigenvector")
graph.set_ylabel("Second eigenvector")
graph.set_zlabel("Third eigenvector")
plt.show()
```
